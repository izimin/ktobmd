# Задание №1. Зимин Илья

```python
from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Мат. функция
def f(x):
  return np.sin(x / 5) * np.exp( x / 10) + 5 * np.exp(-x / 2)
```

```python
# Построение графика функции
xes = np.linspace(1, 15)              # 50 точек по умолчанию
plt.plot(xes, np.vectorize(f)(xes))
```

### 1. Точки 1 и 15 (аппроксимация очень плохая)

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([f(1), f(15)])
```

```python
w = linalg.solve(a, b)
print(w)
```

```python
# Построение первого приближения
plt.plot(xes, w[0] + xes * w[1])
```

### 2. Точки 1, 8, 15 (аппроксимация получше)

```python
a = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b = np.array([f(1), f(8), f(15)])
```

```python
w = linalg.solve(a, b)
print(w)
```

```python
# Построение второго приближения
plt.plot(xes, w[0] + xes * w[1] + xes ** 2 * w[2])
```

### 3. Точки 1, 4, 10, 15 (аппроксимирует хорошо)

```python
a = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
b = np.array([f(1), f(4), f(10), f(15)])
```

```python
w = linalg.solve(a, b)
print(w)
```

```python
# Построение третьего приближения
plt.plot(xes, w[0] + xes * w[1] + xes ** 2 * w[2] + xes ** 3 * w[3])
```

### 4. Запись в файл

```python
file = open('submission-2.txt', 'w')
file.write(f"{w[0]} {w[1]} {w[2]} {w[3]}")
file.close()
```
